# Add Recency Score

```python
pd.qcut(RFM_clean['Recency'], q=5)
```

```python
recency_score = pd.qcut(RFM_clean['Recency'], q=5, labels=range(1, 6))
RFM_clean = RFM_clean.assign(R=recency_score.values)
RFM_clean
```

# Add Frequency Score

```python
pd.qcut(RFM_clean_2['Frequency'], 5, duplicates='drop')
```

```python
RFM_clean['Frequency'].value_counts()
```

frequency_score = pd.qcut(RFM_clean_2['Frequency'], 5,labels=range(1,6), duplicates='drop')
RFM_clean = RFM_clean.assign(F = frequency_score.values)

```python
def frequency_score(x) :
    if x['Frequency'] == 1:
        frequency = 1
    elif x['Frequency'] == 2:
        frequency = 2
    elif x['Frequency'] == 3:
        frequency = 3
    elif x['Frequency'] == 4:
        frequency = 4
    else:
        frequency = 5
    return frequency
RFM_clean['F'] = RFM_clean.apply(frequency_score, axis=1)
RFM_clean
```

# Add Monetary Score

```python
pd.qcut(RFM_clean['Monetary'], q=5)
```

```python
monetary_score = pd.qcut(RFM_clean['Monetary'], q=5, labels=range(1, 6))
RFM_clean = RFM_clean.assign(M=monetary_score.values)
RFM_clean
```

```python
def rfm_score(row):
    return str(row['R']) + str(row['F']) + str(row['M'])
RFM_clean['RFM_Score'] = RFM_clean.apply(rfm_score, axis=1)
RFM_clean
```
