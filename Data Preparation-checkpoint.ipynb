# Cari Recency

```python
left_harga.info()
```

```python
left_harga['TGL_PEMBELIAN'] = left_harga['TGL_PEMBELIAN'].apply(pd.to_datetime)
```

```python
left_harga['Recency'] = (left_harga['TGL_PEMBELIAN'] - pd.Timestamp('now')).dt.days
```

```python
left_harga['Recency'] = left_harga['Recency'].abs()
```

```python
left_harga.sort_values(by='Recency',ascending=False)
```

```python
left_harga
```

```python
left_harga = left_harga.loc[left_harga["Recency"] < 2300]
left_harga
```
